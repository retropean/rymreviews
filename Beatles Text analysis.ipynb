```python
import os  # operating system commands
import re  # regular expressions
import nltk  # draw on the Python natural language toolkit
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from numpy import *  # for array calculations
```

```python
search_word = 'beatles'
```

```python
my_directory = 'C:/Users/00811289/Documents/github/rymreviews'
positive_list = PlaintextCorpusReader(my_directory, 'Hu_Liu_positive_word_list.txt')
negative_list = PlaintextCorpusReader(my_directory, 'Hu_Liu_negative_word_list.txt')
positive_words = positive_list.words()
negative_words = negative_list.words()
```

```python
print(type(negative_words))
print(type(positive_words))
```

```python
positive_scoring={}
for word in positive_words:
    positive_scoring[word] = +1
```

```python
negative_scoring={}
for word in negative_words:
    negative_scoring[word] = -1
```

```python
scoring_dictionary = negative_scoring.copy()
scoring_dictionary.update(positive_scoring)
```

```python
this_file_name = 'C:/Users/00811289/Documents/github/rymreviews/reviews/please_please_me.csv'
with open(this_file_name, 'rt', encoding="utf8") as f:
    please_please_me = f.read()
    
```

```python
textdivider = 10 * 'xxxxxxxx '
```

```python
textdivider
```

```python
please_please_me = please_please_me.replace('<span>', textdivider) # add in that text divider thing
please_please_me = re.sub('<[^<]+?>', '', please_please_me) # remove html
please_please_me = re.sub(r'([^\s\w]|_)+', '', please_please_me) # remove everything except alpha-numeric and spaces
please_please_me = please_please_me.lower()
```

```python
please_please_me.count(search_word) # How many times mentions beatles
```

```python
print(please_please_me.count("standing there")) # check mentions by song...
print(please_please_me.count("twist"))
```

```python
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
```

```python
please_please_me_unfiltered = please_please_me.split()
```

```python
please_please_me_split = []
for w in please_please_me_unfiltered:
    if w not in stop_words:
        please_please_me_split.append(w)
```

```python
please_please_me_split
```

```python
len(please_please_me_split) - please_please_me.count('xxxxxxxx') # 78536 words to analyze
```

```python
please_score = [0] * len(please_please_me_split)
```

```python
for iword in range(len(please_please_me_split)):
    if please_please_me_split[iword] in scoring_dictionary:
        please_score[iword] = scoring_dictionary[please_please_me_split[iword]]
```

```python
sum(please_score) #12553
#len(please_score) # 82456
```

```python
print('Average Sentiment Score:')
print(round(sum(please_score) / (len(please_please_me_split) - please_please_me.count('xxxxxxxx')), 3)) # score / total words
```

```python
def word_sentiment(word):
    blogrelevant = [0] * len(please_please_me_split)  # initialize blog-relevnat indicator
    blogrelevantgroup = [0] * len(please_please_me_split)
    groupcount = 0  

    for iword in range(len(please_please_me_split)):
        if please_please_me_split[iword] == word:
            groupcount = groupcount + 1
            for index in range(max(0,(iword - 5)),min((iword + 5), len(please_please_me_split))):
                blogrelevant[index] = 1
                blogrelevantgroup[index] = groupcount

    return print(round(sum((array(blogrelevant) * array(please_score))) / sum(array(blogrelevant)),3))
```
